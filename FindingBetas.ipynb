```python
# @title
import numpy as np
import matplotlib.pyplot as plt
from google.colab import userdata

# Define the sample size parameter
sample_size_n = 10000

# Create n two-dimensional points
# Generate x-coordinates randomly between 0 and 1
x_coords = np.random.rand(sample_size_n)
# Generate y-coordinates based on the formula y = 0.6x + 0.3 + random number between -0.3 and 0.3
random_noise = -0.3 + (0.3+0.3) * np.random.rand(sample_size_n)
y_coords = 0.6 * x_coords + 0.3 + random_noise

points = np.column_stack((x_coords, y_coords))


# Plot the points
plt.figure(figsize=(6, 6))
plt.scatter(points[:, 0], points[:, 1], s=5)
plt.title(f'Scatter Plot of {sample_size_n} Random Points')
plt.xlabel('X-coordinate')
plt.ylabel('Y-coordinate')
plt.xlim(0, 1)
plt.ylim(0, 1) # Adjust the y-limit back to 0-1
plt.grid(True)
plt.show()
```

## Calculate the means

### Subtask:
Find the mean of the x-coordinates and the mean of the y-coordinates.


**Reasoning**:
Calculate the mean of the x and y coordinates using numpy's mean function.



```python
mean_x = np.mean(points[:, 0])
mean_y = np.mean(points[:, 1])
```

## Calculate deviations

### Subtask:
Subtract the mean of x from each x-coordinate and the mean of y from each y-coordinate.


**Reasoning**:
Subtract the means from the respective coordinates to calculate deviations.



```python
x_deviations = points[:, 0] - mean_x
y_deviations = points[:, 1] - mean_y
```

## Calculate the slope (m)

### Subtask:
Use the formula m = dot_product((x - mean_x), (y - mean_y)) / dot_product((x - mean_x), (x - mean_x)) to calculate the slope of the linear regression line.


**Reasoning**:
Calculate the slope of the linear regression line using the dot products of the deviations.



```python
numerator = np.dot(x_deviations, y_deviations)
denominator = np.dot(x_deviations, x_deviations)
beta1 = numerator / denominator
```

## Calculate the intercept (b)

### Subtask:

```python
beta0 = mean_y - beta1 * mean_x
```

## Calculate predicted y-values

### Subtask:
Use the calculated `beta0` and `beta1` to predict the y-values for each x-coordinate.


**Reasoning**:
Calculate the predicted y-values using the linear regression equation with the calculated beta0 and beta1 and store them in a new variable y_predicted.



```python
y_predicted = beta1 * points[:, 0] + beta0
```

## Calculate the total sum of squares (sst)

### Subtask:
Calculate the sum of the squared differences between each original y-value and the mean of the y-values.


**Reasoning**:
Calculate the sum of the squared differences between each original y-value and the mean of the y-values.



```python
sst = np.sum((points[:, 1] - mean_y)**2)
```

## Calculate the residual sum of squares (ssr)

### Subtask:
Calculate the sum of the squared differences between each original y-value and its predicted y-value.


**Reasoning**:
Calculate the sum of the squared differences between the original y-values and the predicted y-values using numpy's sum function.



```python
ssr = np.sum((points[:, 1] - y_predicted)**2)
```

## Calculate r-squared

### Subtask:
Use the formula R-squared = 1 - (SSR / SST) to calculate the R-squared value.


**Reasoning**:
Calculate the R-squared value using the formula R-squared = 1 - (SSR / SST).



```python
r_squared = 1 - (ssr / sst)
```

## Output the r-squared value

### Subtask:
Display the calculated R-squared value.


**Reasoning**:
Display the calculated R-squared value.



```python
print(f"The R-squared value is: {r_squared}")
```
